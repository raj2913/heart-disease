## Modeling

```python
#Here we are dividing the data into two variables X includes the values which are used to predict the heart disease and Y includes whether the person is having heart disease or not
X=hd.drop('target',axis=1)
Y=hd.target
np.random.seed(42)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#Checking the count of training dataset
len(X_train)
```

```python
#Checking the count of testing dataset
len(Y_test)
```

```python
#Checking the models that we think are suitable for the data 
models={"Logistic":LogisticRegression(),
       "Neighbours":KNeighborsClassifier(),
       "RandomForest":RandomForestClassifier()}
def fit_and_score(models,X_train,X_test,Y_train,Y_test):
    model_scores={}
    for names,model in models.items():
        model.fit(X_train,Y_train)
        model_scores[names]=model.score(X_test,Y_test)
    return model_scores
```

```python
#Here we are fitting the data fitting includes training the data using the training data which we have split(X_train,Y_train) and then score includes testing the model and checking the accuracy of the model we are passing both parameter in score(X_test and y_test)
#The trained model based on the previous training dataset predict the Y_test values using X_test data and cross verify with the original Y_test
#score=correctly predicted/total number of testing data(61)
scores=fit_and_score(models=models,
                    X_train=X_train,
                    X_test=X_test,
                    Y_train=Y_train,
                    Y_test=Y_test)
```

```python
#Scores of all the models
scores
```

```python
#We are having a dataframe for the above scores
result=pd.DataFrame(scores,index=["accuracy"])
```

```python
#Vizualizing the dataframe
result.T.plot(kind='bar',figsize=(5,5))
plt.title("Model Comparision");
```

Conclusion:We should move on with the logisticregression model and RandomForest model as it is having more accuracy

# Tuning Hyperparameters

```python
# Creating a hyperparameter grid for LogisticRegression
log_reg_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}

# Creating a hyperparameter grid for RandomForestClassifier
rf_grid = {"n_estimators": np.arange(10, 1000, 50),
           "max_depth": [None, 3, 5, 10],
           "min_samples_split": np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(1, 20, 2)}
```

```python
gs_log_reg_grid=GridSearchCV(LogisticRegression(),param_grid=log_reg_grid,cv=5,verbose=True)
gs_log_reg_grid.fit(X_train,Y_train)
```

```python
gs_log_reg_grid.best_params_
```

```python
G_log=LogisticRegression(C=0.23357214690901212,solver='liblinear')
```

```python
G_log.fit(X_train,Y_train)
G_log.score(X_test,Y_test)
```

```python
np.random.seed(42)
rs_rf_grid=RandomizedSearchCV(RandomForestClassifier(),param_distributions=rf_grid,cv=5,n_iter=10,verbose=True,)
rs_rf_grid.fit(X_train,Y_train)
```

```python
rs_rf_grid.best_params_
```

```python
R_ran=RandomForestClassifier(n_estimators=310,min_samples_split=2,min_samples_leaf=19,max_depth=5)
R_ran.fit(X_train,Y_train)
R_ran.score(X_test,Y_test)
```
